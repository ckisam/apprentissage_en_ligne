# Projet d'apprentissage en ligne

Ce projet s'appuie sur un article de Elad Hazan, Adam Kalai, Satyen Kale et Amit Agarwal « Logarithmic Regret Algorithms for Online Convex Optimization ». Les auteurs y présentent trois algorithmes permettant de faire de l'optimisation convexe en ligne, en garantissant sous certaines conditions un regret logarithmique à la date $T$. Avec un focntion de perte convexe $f_t$ et une décision $x_t$ à la date $t$, le regret est défini par:
$$ Regret : R_T = \sum\limits_{t=1}^T f_t(x_t) - \underset{x \in K}{\text{min}} \sum\limits_{t=1}^T f_t(x)$$


Les trois algorithmes étudiés dans cet article sont :
- **Online Gradient Descent**
- **Online Newton Step**
- **Exponentially Weighted Online Optimization**

## Chargement des données

On exploite un jeu de données de sélection de portefeuille. Il s'agit de l'indice boursier S&P 500 pour 490 sociétés, durant 1000 jours de marché de la période 2001 à 2005. Le jeu de données a été obtenu à ce [lien](http://ocobook.cs.princeton.edu/links.htm) au format Matlab puis converti au format CSV.

On cherche donc, au cours du temps, à obtenir le meilleur portefeuille parmi les $n$ = 490 actions.

```python
data = pd.read_csv("data.csv")
print("Dimensions du DF 'data' : {}".format(data.shape))
data.head()
```

A chaque étape $t$, on choisit un portefeuille d'actions $x_t \in \mathbb{R}^{n}$. La valeur des indices $a_t$ à chaque période est donnée par l'accroissement $c_t = \frac{a_{t+1}}{a_t}$. 

```python
prixFull = data.copy().iloc[1:data.shape[0],]
for i in range(0, prixFull.shape[0]) :
    prixFull.iloc[i, ] = data.iloc[i+1, ] / data.iloc[i, ]
prixFull = prixFull.reset_index(drop = True)
print("Dimensions du DF 'prixFull' : {}".format(prixFull.shape))
prixFull.head()
```

Pour des raisons de performance, on choisit de se limiter à un choix de portefeuille parmi les $n$ = 10 premières actions.

```python
### Réduction de la dimension à 10 actions
prix = prixFull.iloc[:, 0:10]
prix.head()
```

## Définition d'une fonctionde perte

L'investisseur subit une perte logarithmique $\mathcal{L}(x_t) = -\log(x_t^Tc_t)$.

```python
def Loss(x, c):
    return -math.log((x * c).sum())

def Gradient(x, c):
    scalarProd = (x * c).sum()
    return -c / scalarProd
```

On choisit de normer l'ensemble des placements à $1$. Les décisions possibles appartiennent donc au simplexe $K = \{x \in \mathbb{R}_+^{n} \text{ tq } x^Tu = 1\}$ où le vecteur $u$ n'est composé que de $1$. On définit la fonction de projection sur cet ensemble.

```python
def projectionSimplexe(y):
    d = len(y)
    u = np.ones(d)
    x = cvx.Variable(d)
    obj = cvx.Minimize(cvx.sum_squares(x - y))
    constr = [x >= 0, u*x == 1]
    prob = cvx.Problem(obj, constr)
    prob.solve()
 
    return np.array(x.value).squeeze().tolist()
```

## Calcul du regret

Pour évaluer la performance de nos algorithmes, nous calculons à chaque étape l'erreur logarithmique. Pour obtenir le regret, il est donc nécessaire d'ajouter le terme associé à la meilleure sélection de portefeuille sur l'ensemble des périodes : $\underset{x \in K}{\text{min}} \sum\limits_{t=1}^T f_t(x)$

On passe par un programme d'optimisation sous contrainte.

```python
def objective(x, df):
    return df.apply(func = lambda y: Loss(x, y), axis = 1).sum()

def simplexConstraint(x):
    unitVect = np.ones(len(x))
    return (unitVect * x).sum() - 1

def bestPrediction(df):
    p = df.shape[1]
    startMin = np.ones(p)/p
    resOpt = scipy.optimize.minimize(fun = objective, x0 = startMin, args = (df),
                                  constraints = [
                                      {'type' : 'eq', 'fun' : simplexConstraint},
                                      {'type' : 'ineq', 'fun' : lambda x: x}
                                  ]).x
    res = dict()
    res['x'] = resOpt
    res['value'] = objective(resOpt, df)
    return res
```

Le calcul du meilleur portefeuille est un peu long. Nous le calculons toutes les 100 journées de marché pour avoir une idée de l'évolution.

```python
dayRef = [98, 198, 298, 398, 498, 598, 698, 798, 898, 998]
start = time.time()
bestPredList = pd.Series(dayRef).apply(func = lambda i: bestPrediction(prix.iloc[0:i, ]))
print("Calcul des meilleurs prédictions effectuées en {} secondes".format(time.time() - start))
```

```python
def plotRegret(error, dayRef = dayRef, bestPredList = bestPredList):
    cumError = pd.Series(dayRef).apply(func = lambda i: sum(error[0:i]))
    bestPredValue = bestPredList.apply(func = lambda x: x['value'])
    res = cumError - bestPredValue
    res.index = dayRef
    res.plot()
    plt.title("Regret de l'algorithme")
```

## Algorithme de la descente de gradient en ligne

Descente de gradient dans le cadre ** en ligne**. 

Pas : $\eta_1, ..., \eta_T$ avec $\eta_t=\frac{1}{H(t+1)}$

>  Etape 1: On choisit arbitrirment $x_1$

> Etape 2: Pour tout $t>1$: on 
>> $x_{t+1} = \Pi_K(x_t - \eta_{t+1} \nabla f_t (x_t))$



 >>avec $\Pi_K(y)$: une projection sur $K$, 
$\Pi_K(y) = \underset{x \in K}{\text{argmin}} ||x-y||$

```python
def onlineGradientDescent(df, init, H = 0.01):  
    x_t_normalise= projectionSimplexe(init) 
    new_x_t=[x_t_normalise]
    loss = []
    prediction= []
    for i in range(df.shape[0]-1):
        c = df.iloc[i, ]
        eta_t=1/(H*(i+1))
        new_x_t += [projectionSimplexe((new_x_t[i] - eta_t * Gradient(new_x_t[i], c)).tolist())]
        c = df.iloc[i+1, ]
        loss += [Loss(new_x_t[i+1], c) ]
    res = dict()
    res["prediction"] = new_x_t
    res["loss"] = loss
    return(res)

ogdResult = onlineGradientDescent(prix, np.ones(prix.shape[1])/prix.shape[1])
```

```python
ogdError = ogdResult["loss"]

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
pd.Series(ogdError).plot()
plt.title("Perte logarithmique au cours du temps")
plt.subplot(1,3,2)
pd.Series(np.cumsum(ogdError)).plot()
plt.title("Perte cumulée au cours du temps")
plt.subplot(1,3,3)
plotRegret(ogdError)
plt.show()
```

## Algorithme de Newton en ligne

Descente de gradient dans le cadre ** en ligne**. 


>  Etape 1: On choisit arbitrirment $x_1$

> Etape 2: Pour tout $t>1$: on a

>> $$\Delta_{t-1}=\Delta f_{t-1}(x_{t-1})$$

>> $$A_{t-1}= \sum_{\tau=1}^{t-1}\Delta_{\tau}\Delta_{\tau}^T$$


>> $$b_{t-1}= \sum_{\tau=1}^{t-1}\Delta_{\tau}\Delta_{\tau}^T x_\tau- \frac{1}{\beta}\Delta_{\tau}$$



>> $$x_{t+1} = \Pi_K^{A_{t-1}}(A_{t-1}b_{t-1})$$



 >>avec $\Pi_K^{A_{t-1}}(y)$: une projection sur $K$, 
$$\Pi_K^{A_{t-1}}(y) = \underset{x \in K}{\text{argmin}} (x-y)^T A_{t-1}(x-y)$$

```python
def projectionSelonMatrice(A, y) :
    d = len(y)
    u = np.ones(d)
    x = cvx.Variable(d)
    obj = cvx.Minimize(cvx.quad_form(x - y, A))
    constr = [x >= 0, u*x == 1]
    prob = cvx.Problem(obj, constr)
    prob.solve()
 
    return np.array(x.value).squeeze()

def onlineNewtonStep(df, init, beta = 0.01):
    start = time.time()
    print("--- ALGORITHME ONLINE NEWTON STEP")
    n, p = df.shape
    portfolio = [ init ]
    loss = []
    A = np.zeros((p, p))
    b = np.zeros(p)
    print("--- Nombre d'itérations : {}".format(n))
    it10pct = n // 10
    for i in range(n-1):
        c = df.iloc[i, ]
        delta = Gradient(portfolio[i], c).values.reshape(p, 1)
        Anew = np.dot(delta, np.transpose(delta))
        delta = np.transpose(delta)[0]
        bNew = np.dot(Anew, portfolio[i]) - delta / beta
        A += Anew
        b += bNew
        portfolio += [ projectionSelonMatrice(A, np.dot(np.linalg.pinv(A), b)) ]
        c = df.iloc[i+1, ]
        loss += [ Loss(portfolio[i+1], c) ]
        if i % it10pct == 0:
            print("----- {}% de l'algorithme".format((i // it10pct) * 10))
        
    print(" --- FIN DE L'ALGORITHME")
    print(" --- Temps écoulé : {} secondes".format(time.time() - start))
    res = dict()
    res["prediction"] = portfolio
    res["loss"] = loss
    return(res)

onsResult = onlineNewtonStep(prix, np.ones(prix.shape[1])/prix.shape[1])
```

```python
onsError = onsResult["loss"]

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
pd.Series(onsError).plot()
plt.title("Perte logarithmique au cours du temps")
plt.subplot(1,3,2)
pd.Series(np.cumsum(onsError)).plot()
plt.title("Perte cumulée au cours du temps")
plt.subplot(1,3,3)
plotRegret(onsError)
plt.show()
```

## Optimisation en ligne avec poids exponentiels

On commence par introduire une marche aléatoire permettant d'échantillonner selon $w$.

Généralisation de l'algorithme $EWA$ au cas continu.


 >> Etape 1: Poids exponentiels $w_{t+1}(x) = e^{-\alpha\sum\limits_{\tau = 1}^{t} f_{\tau}(x)} = \prod\limits_{\tau = 1}^{t} h_{\tau}(x)$
 
 


>> Etape 2:Mise à jour $x_{t+1} = \frac{\int_K x w_{t+1}(x)dx}{\int_K w_{t+1}(x)dx}$




```python
import random

def computeWeight(alpha, df, t, x):
    subDf = df.iloc[0:t,]
    prod = np.dot(subDf, x)
    # print("sous-DF : {} - x : {} - prod : {}".format(subDf.shape, x.shape, prod.shape))
    res = math.exp(-alpha * prod.sum())
    return(res)

def randomWalk(alpha, df, t, delta):
    n = df.shape[1]
    x = np.ones(n) / n
    currentLkh = computeWeight(alpha, df, t, x)
    nbIt = 10 * df.shape[1]
    
    for i in range(nbIt):
        j = np.random.choice([k for k in range(n-1)])
        X = np.random.choice([-1,1])
        y = x.copy()
        y[j] = y[j] + X*delta
        y[n-1] = y[n-1] - X*delta
        newLkh = computeWeight(alpha, df, t, y)
        proba = min(1, currentLkh / newLkh)
        if np.random.binomial(1, proba) == 1 :
            x = y.copy()
            currentLkh = newLkh
    return x

def exponentiallyWeighted(df, init, alpha = 0.01, delta = 0.01):
    start = time.time()
    print("--- ALGORITHME EXPONENTIALLY WEIGHTED")
    n = df.shape[0]
    portfolio = [ init ]
    loss = []
    print("--- Nombre d'itérations : {}".format(n))
    it10pct = n // 10
    for i in range(n-1):
        c = df.iloc[i, ]
        
        portfolio += [ randomWalk(alpha, df, i, delta) ]
        c = df.iloc[i+1, ]
        loss += [ Loss(portfolio[i+1], c) ]
        if i % it10pct == 0:
            print("----- {}% de l'algorithme".format((i // it10pct) * 10))
        
    print(" --- FIN DE L'ALGORITHME")
    print(" --- Temps écoulé : {} secondes".format(time.time() - start))
    res = dict()
    res["prediction"] = portfolio
    res["loss"] = loss
    return(res)
    

ewooResult = exponentiallyWeighted(prix, np.ones(prix.shape[1])/prix.shape[1])
```

```python
ewooError = ewooResult["loss"]

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
pd.Series(ewooError).plot()
plt.title("Perte logarithmique au cours du temps")
plt.subplot(1,3,2)
pd.Series(np.cumsum(ewooError)).plot()
plt.title("Perte cumulée au cours du temps")
plt.subplot(1,3,3)
plotRegret(ewooError)
plt.show()
```

## Follow the approximate leader (FTAL)

```python
def f_tilde(x, x_t, c, beta):
    scalarProd = (Gradient(x_t, c) * (x - x_t)).sum()
    return(Loss(x_t,c) + scalarProd + (beta / 2) * (scalarProd**2))

def f_tilde_sum (x, predList, df, beta):
    res = 0
    for i in range(len(predList)):
        c = df.iloc[i, ]
        res += f_tilde(x, predList[i], c, beta)
    return res

def simplexConstraint(x):
    unitVect = np.ones(len(x))
    return (unitVect * x).sum() - 1

def followApproximateLeader(df, init, beta = 0.01):
    start = time.time()
    print("--- ALGORITHME 'FOLLOW THE APPROXIMATE LEADER'")
    
    n, p = df.shape
    predList = [ init ]
    lossList = []
    
    print("--- Nombre d'itérations : {}".format(n))
    it10pct = n // 10
    
    for i in range(n-1):
        startMin = np.ones(p) / p
        newPred = scipy.optimize.minimize(fun = f_tilde_sum, x0 = startMin, args = (predList, df, beta),
                                         constraints = [
                                             {'type' : 'eq', 'fun' : simplexConstraint},
                                             {'type' : 'ineq', 'fun' : lambda x: x}
                                         ]).x
        predList += [ newPred ]
        c = df.iloc[i, ]
        lossList += [ Loss(newPred, c) ]
        
        if i % it10pct == 0:
            print("----- {}% de l'algorithme".format((i // it10pct) * 10))
        
    print(" --- FIN DE L'ALGORITHME")
    print(" --- Temps écoulé : {} secondes".format(time.time() - start))
    res = dict()
    res["prediction"] = portfolio
    res["loss"] = loss
    return(res)

'''
ftalResult = followApproximateLeader(prix, np.ones(prix.shape[1])/prix.shape[1])
ftalError = ftalResult["loss"]

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
pd.Series(ftalError).plot()
plt.title("Perte logarithmique au cours du temps")
plt.subplot(1,2,2)
pd.Series(np.cumsum(ftalError)).plot()
plt.title("Perte cumulée au cours du temps")
plt.show()
'''
```
